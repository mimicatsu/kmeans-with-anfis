```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster, metrics
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
```

```python
path = "../bankruptcy_HW1_train.csv"
data = pd.read_csv(path).drop(["Unnamed: 0"],axis=1)
```

```python
data
```

```python
data_test = pd.read_csv("../bankruptcy_HW1_test.csv").drop(["Unnamed: 0"],axis=1)
```

```python
data_test
```

```python
#get the y and y_test
y = data["Bankrupt?"]
y_test = data_test["Bankrupt?"]
```

```python
#read the data which output from matlab
#the fis is using kmeans
y_train_pred = pd.read_excel("bankruptcy_train_ANFIS_Output.xlsx")
y_test_pred = pd.read_excel("bankruptcy_test_ANFIS_Output.xlsx")
```

```python
y_train_pred 
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
print("Accuracy_train:",accuracy_score(y, y_train_pred))
print(classification_report(y, y_train_pred))
print("Accuracy_test:",accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

```python
#read the data which output from matlab
#the fis is using Gird partition
y_train_pred = pd.read_excel("bankruptcy_train_Grid_Output.xlsx")
y_test_pred = pd.read_excel("bankruptcy_test_Grid_Output.xlsx")
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
print("Accuracy_train:",accuracy_score(y, y_train_pred))
print(classification_report(y, y_train_pred))
print("Accuracy_test:",accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```
